```python
import spacy # contains model which we are going to use
from spacy import displacy # to display along with tag
import pandas as pd
```

```python
df=pd.read_csv("/content/apple.txt",on_bad_lines="skip")
df
```

```python
# how to convert these tabular sentences to paragraph for a corpus
df['x'] # x is a column name
```

```python
# we want to join all these sentences
nlp=spacy.load("en_core_web_sm")  # model name is an  en_core_web_sm, nlp will now have entire NER model
# en_core_web_sm is a small english pipe line on written web text , that includes vocabu;ary,syntax,and entities
```

```python
nlp.pipe_names # shows different models are present inside en_core-web_sm model
```

```python
# we want ner (name entity recognition)

doc= nlp("Mr.elon is going to acquire twitter icn of $45 billion in 2023")
```

```python
doc.ents   #ents -will display all entities
```

```python
for i in doc.ents:
  print(i,'|',i.label_)
```

SpaCy library is not 100% precise

```python
# to display graphical representation for tags
displacy.render(doc,style='ent',jupyter=True) # ent-entity, In Jupyter of serve() try render()
```

```python
# joining all sentences of dataset
txt=" ".join(df['x']) # with space join all sentences in x
txt
```

```python
x=nlp(txt)
```

```python
displacy.render(x,style='ent') # 70 to 80 accurate model
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
```

A word cloud is a visualization technique for text data where the most frequentword is shown in biggest font size

you can create customized word cloud in python

```python
wc=WordCloud(background_color='white').generate(txt)
plt.imshow(wc)#image show
plt.axis("off")
plt.title("wordcloud",size=25)

# bigger the name of word more the frequency of that word in the corpus
# font of apple, laptop is huge i.e., it is occuring many times in text
# word cloud is important when we are analysing the data when we build Naive Bais text classification.
# this is how we can create word cloud
```
